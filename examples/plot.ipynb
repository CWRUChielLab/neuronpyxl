```python
def freq(t, v, thresh=1e-3):
    indices = np.where((v[:-1] < -thresh) & (v[1:] > thresh))[0]
    return t[indices]
```

```python
v1 = data["V_B8"]
t = data["t"]

fig,ax = plt.subplots(1,1,figsize=(10,5))
plt.plot(t,v1)
zeros = freq(t,v1)
yzeros = np.zeros_like(zeros)
plt.scatter(zeros,yzeros)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=0, scale=1, size=1000)

fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)

ax.hist(data, bins=30, color="#1f77b4", edgecolor='white', alpha=0.8)

# Styling
ax.set_title("Histogram of Data", fontsize=14, fontweight='bold')
ax.set_xlabel("Value", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
