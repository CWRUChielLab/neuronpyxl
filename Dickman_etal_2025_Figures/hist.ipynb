```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
filepath = "./Data/nostim_data"
file = pd.HDFStore(os.path.join(filepath,"nostim_data.h5"))
snnap_data = pd.read_csv(os.path.join(filepath,"excitability.smu.out"),sep="\t").dropna(axis=1)
snnap_data.columns = ["t","V_B4","I1","I2","I3","I4","I5","I6","I7"]
data = file["membrane"]
```

```python
v = data["V_B4"].to_numpy()
snnap_mask = snnap_data["t"] > 10
vsnnap = snnap_data["V_B4"][snnap_mask].to_numpy()
```

```python
# Histogram with KDE
sns.histplot(v, kde=True, bins=30, color="red", alpha=0.5, edgecolor="none",stat="density")
sns.histplot(vsnnap, kde=True, bins=30, color="blue", alpha=0.5, edgecolor="none",stat="density")
```
